```python
# Importing Pandas for Data Cleaning and Manipulation

import pandas as pd
```

```python
#Importing Raw Data Set

df = pd.read_csv('SuperStore_raw_dataset.csv')
```

```python
#Checking the first few rows

print(df.head())
```

```python
#Checking Dataset Info

df.info()
```

```python
#Check Shape of dataset

print(df.shape)
```

```python
#Convert the datatypes for order date and ship date to datetime data type

df['Order Date'] = pd.to_datetime(df['Order Date'], errors = 'coerce', dayfirst=True)

df['Ship Date'] = pd.to_datetime(df['Ship Date'], errors = 'coerce', dayfirst=True)
```

```python
#Check datatype conversion

df.info()
```

```python
#Check if any values failed to parse

failed_order = df['Order Date'].isna().sum()
failed_ship = df['Ship Date'].isna().sum()
print(failed_order)
print(failed_ship)
```

```python
#Create Delivery days column from ship date and order date

df['delivery_days'] = (df['Ship Date'] - df['Order Date']).dt.days
```

```python
#Check if the delivery_days column was created

df
```

```python
#Extract date based features

df['order_year'] = df['Order Date'].dt.year
df['order_month'] = df['Order Date'].dt.month
df['order_month_name'] = df['Order Date'].dt.strftime('%b')
df['order_quarter'] = df['Order Date'].dt.to_period('Q').astype(str)
df['order_day'] = df['Order Date'].dt.day_name()

```

```python
#Check the extracted date features

df
```

```python
#Check for missing postal codes

df['Postal Code'].isna().sum()
```

```python
#Filling the missing postal codes with mode 

df['Postal Code'].fillna(df['Postal Code'].mode()[0], inplace=True)
```

```python
#Filling the missing postal codes with mode without using inplace

df['Postal Code'] = df['Postal Code'].fillna(df['Postal Code'].mode()[0])
```

```python
df['Postal Code'].isna().sum()
```

```python
df.info()
```

```python
#Drop Country Column as it consists of only one value

df= df.drop(columns='Country') 
```

```python
#Check if row id values are unique

df['Row ID'].is_unique
```

```python
#Check total duplicate rows

df.duplicated().sum()
```

```python
#Check text columns in dataset

df.select_dtypes(include='object').columns
```

```python
#Clean all text columns

for col in df.select_dtypes(include='object').columns:
    df[col] = (
        df[col]
        .astype(str)                # make sure everything is string
        .str.strip()                # remove leading/trailing spaces
        .str.replace('\xa0', ' ', regex=False)  # remove non-breaking spaces
        .str.lower()                # convert to lowercase
    )
```

```python
#Make all column names sql friendly

df.columns = (
    df.columns
    .str.strip()                # remove leading/trailing spaces
    .str.lower()                # convert to lowercase
    .str.replace(' ', '_')      # replace spaces with underscores
    .str.replace('-', '_')      # replace hyphens with underscores
    .str.replace('/', '_')      # replace slashes with underscores
)
```

```python
df['type_of_day'] = df['order_day'].apply(
    lambda x: 'Weekend' if x in ['Saturday', 'Sunday'] else 'Weekday'
)
```

```python
#Save Cleaned dataset

df.to_csv('Superstore_cleaned_final.csv', index=False)
```
